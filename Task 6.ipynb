```python
import pandas as pd
# Replace 'your_file.csv' with the path to your CSV file
df = pd.read_csv('/content/dataset.csv')
```

```python
df['order_id'] = pd.to_numeric(df['order_id'], errors='coerce')
df.dropna(subset=['order_id'], inplace=True)
df['order_id'] = df['order_id'].astype(int)
```

```python
df = df[df['product_id'] != 0]
```

```python
df['amount'] = df['amount'].apply(lambda x: 1500 if x > 1500 else x)
```

```python
df.dropna(subset=['status'], inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
from sqlalchemy import create_engine, Column, Integer, String, Float, Table, MetaData

# Create an engine to the SQLite database (or another DB of your choice)
engine = create_engine('sqlite:///final_data.db', echo=True)

# Define the table schema using SQLAlchemy's MetaData
metadata = MetaData()
orders = Table('orders', metadata,
    Column('order_id', Integer, primary_key=True),
    Column('product_id', Integer),
    Column('amount', Float),
    Column('status', String)
)

# Create the table
metadata.create_all(engine)
```

```python
# Load the data into the 'orders' table
df.to_sql('orders', con=engine, if_exists='replace', index=False)
```
